```python
# import requests
# from bs4 import BeautifulSoup

transcript_url = 'https://adventuretime.fandom.com/wiki/Slumber_Party_Panic/Transcript'
response = requests.get(transcript_url)
soup = BeautifulSoup(response.text, 'html.parser')

```

```python
soup.prettify()
```

```python
# Maybe not the best way to navigate …
h2_transcript = soup.find('span', id='Transcript').parent

last_elem = h2_transcript

for i in range(10):
    curr_line = last_elem.next_sibling.next_sibling
    print(curr_line)
    # Move to the next line
    last_elem = curr_line
```

```python
transcript_lines = soup.find_all('dl')

lines = []
for line in transcript_lines:
    line_str = line.find('dd').text.strip()
    
    print(line_str)
    lines.append(line_str)
    
# Alternative
# line_str = line.dd.text
# line_str
```

```python
transcript_lines = soup.find_all('dl')

lines_with_info = []
for line in transcript_lines:
    line_str = line.find('dd').text.strip()
    
    # Test if it's extra info (starts with '[') → 'extra_info'
    if(line_str[0] == '['):
        person = 'extra_info'
    # If not, assume a person → that character's name
    else:
        split_info = line_str.split(':')
        person = split_info[0]
        line_str = split_info[1].strip()
        
    lines_with_info.append((person,line_str))
    
```

```python
n_arr = np.array(lines_with_info)

list(n_arr[:,0])
```

```python
df = pd.DataFrame(n_arr,columns=['speaking','spoken'])
df.speaking.value_counts()
```
